```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

def google_news_scraper(keywords, target_total=500):
    options = Options()
    options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36')
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option('useAutomationExtension', False)
    
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    
    final_results = []
    seen_links = set()

    for kw in keywords:
        if len(final_results) >= target_total:
            break
            
        print(f"\nğŸš€ í‚¤ì›Œë“œ ê²€ìƒ‰: [{kw}]")
        
        for page in range(0, 50): # ìµœëŒ€ 50í˜ì´ì§€ íƒìƒ‰
            if len(final_results) >= target_total:
                break
                
            start_index = page * 10
            url = f"https://www.google.com/search?q={kw}&tbm=nws&start={start_index}"
            driver.get(url)
            time.sleep(3.5) # êµ¬ê¸€ íƒì§€ ë°©ì§€ë¥¼ ìœ„í•´ ì•½ê°„ ë„‰ë„‰íˆ ëŒ€ê¸°

            # ë‰´ìŠ¤ ì¹´ë“œ ì „ì²´ë¥¼ ê°ì‹¸ëŠ” ë¶€ëª¨ ìš”ì†Œë“¤ì„ ì°¾ìŠµë‹ˆë‹¤.
            # ì œê³µí•˜ì‹  ê²½ë¡œì˜ ê³µí†µ ë¶„ëª¨ì¸ a íƒœê·¸ì™€ ë‚´ë¶€ ì»¨í…Œì´ë„ˆ ìœ„ì£¼ë¡œ íƒìƒ‰í•©ë‹ˆë‹¤.
            articles = driver.find_elements(By.CSS_SELECTOR, 'div#rso > div > div > div')

            if not articles:
                print(f"   - {page+1}í˜ì´ì§€ì—ì„œ ë‰´ìŠ¤ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
                break

            for article in articles:
                try:
                    # 1. ë§í¬ ê²½ë¡œ: div > div > a
                    link_el = article.find_element(By.CSS_SELECTOR, 'a')
                    link = link_el.get_attribute('href')
                    
                    if not link or "google.com" in link or link in seen_links:
                        continue
                    
                    # 2. ì œëª© ê²½ë¡œ (ì œê³µí•´ì£¼ì‹  í´ë˜ìŠ¤: n0jPhd)
                    try:
                        title = link_el.find_element(By.CSS_SELECTOR, '.n0jPhd').text
                    except:
                        title = "ì œëª© ì—†ìŒ"

                    # 3. ë¯¸ë¦¬ë³´ê¸° ê²½ë¡œ (ì œê³µí•´ì£¼ì‹  í´ë˜ìŠ¤: UqSP2b)
                    try:
                        snippet = link_el.find_element(By.CSS_SELECTOR, '.UqSP2b').text
                    except:
                        snippet = ""

                    # 4. ë‚ ì§œ ê²½ë¡œ (ì œê³µí•´ì£¼ì‹  ìœ„ì¹˜ ê¸°ë°˜: span íƒœê·¸)
                    try:
                        # ë³´í†µ ë‚ ì§œëŠ” .OSrXXb í´ë˜ìŠ¤ë‚˜ íŠ¹ì • spanì— ìœ„ì¹˜
                        date = link_el.find_element(By.CSS_SELECTOR, '.OSrXXb, span:last-child').text
                    except:
                        date = "ë‚ ì§œ ë¯¸ìƒ"

                    final_results.append({
                        "í‚¤ì›Œë“œ": kw,
                        "ì œëª©": title.strip(),
                        "ë‚ ì§œ": date.strip(),
                        "ë¯¸ë¦¬ë³´ê¸°": snippet.strip(),
                        "ë§í¬": link
                    })
                    seen_links.add(link)

                    if len(final_results) >= target_total:
                        break
                except:
                    continue
            
            print(f"   - í˜„ì¬ ìˆ˜ì§‘ëŸ‰: {len(final_results)}ê°œ")

    driver.quit()
    return final_results

# --- ì‹¤í–‰ ë° CSV ì €ì¥ ---
search_list = ["ì² ê°• ì „ê¸°", "ì‚°ì—… ì „ê¸°", "ì „ë ¥ íš¨ìœ¨", "ê³µì¥ ì „ê¸°"]
data = google_news_scraper(search_list, 500)

if data:
    df = pd.DataFrame(data)
    # ì¤‘ë³µ ì œê±° (í˜¹ì‹œ ëª¨ë¥¼ ì¤‘ë³µ ë°©ì§€)
    df = df.drop_duplicates(subset=['ë§í¬'])
    df.to_csv("google_news_final.csv", index=False, encoding="utf-8-sig")
    print(f"\nâœ… ì™„ë£Œ! 'google_news_final.csv'ì— {len(df)}ê±´ ì €ì¥ë¨")
else:
    print("\nâŒ ìˆ˜ì§‘ëœ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤. ì…€ë ‰í„°ë¥¼ ë‹¤ì‹œ ì ê²€í•´ì•¼ í•©ë‹ˆë‹¤.")
```
